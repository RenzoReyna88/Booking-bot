```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
apartamentos_cdadmndza = pd.read_csv("Datos-Apartamentos-Mndza/Apartamentos-CdadMendoza.csv", sep=',', encoding='utf-8')
apartamentos_lujandecuyo = pd.read_csv("Datos-Apartamentos-Mndza/Apartamentos-LujanDeCuyo.csv", sep=',', encoding='utf-8')
apartamentos_malargue = pd.read_csv("Datos-Apartamentos-Mndza/Apartamentos-Malargüe.csv", sep=',', encoding='utf-8')
apartamentos_sanrafael = pd.read_csv("Datos-Apartamentos-Mndza/Apartamentos-SanRafael.csv", sep=',', encoding='utf-8')
apartamentos_tunuyan = pd.read_csv("Datos-Apartamentos-Mndza/Apartamentos-Tunuyan.csv", sep=',', encoding='utf-8')
apartamentos_tupungato = pd.read_csv("Datos-Apartamentos-Mndza/Apartamentos-Tupungato.csv", sep=',', encoding='utf-8')
apartamentos_uspallata = pd.read_csv("Datos-Apartamentos-Mndza/Apartamentos-Uspallata.csv", sep=',', encoding='utf-8')
```

```python
apartamentos_mndza = pd.concat([apartamentos_cdadmndza, apartamentos_lujandecuyo, 
                                apartamentos_malargue, apartamentos_sanrafael,
                                apartamentos_tunuyan, apartamentos_tupungato, 
                                apartamentos_uspallata
                                ], ignore_index=True)
```

```python
df_apartamentos_mndza = pd.DataFrame(apartamentos_mndza)

df_apartamentos_mndza
```

```python
# Transformar valores string a tipo numéricos de la columna Precio Apartamento
df_apartamentos_mndza['Precio alojamiento'] = df_apartamentos_mndza['Precio alojamiento'].str.replace('$', '').str.replace('.', '').astype(float)


# Convvertir nombre de apartamentos en letra minúscula 
df_apartamentos_mndza['Nombre alojamiento'] = df_apartamentos_mndza['Nombre alojamiento'].str.lower()
```

```python
df_apartamentos_mndza
```

```python
df_apartamentos_mndza['Tipo alojamiento'] = "Departamento"
```

```python
df_apartamentos_mndza.info()
```

```python
df_apartamentos_mndza.describe()
```

```python
print("Valores nulos encontrados:", np.sum(df_apartamentos_mndza.isna()))

print("Valores repetidos encontrados:", np.sum(df_apartamentos_mndza.duplicated()))
```

```python
df_apartamentos_mndza.drop_duplicates(inplace=True)

print("Valores repetidos encontrados:", np.sum(df_apartamentos_mndza.duplicated()))
```
